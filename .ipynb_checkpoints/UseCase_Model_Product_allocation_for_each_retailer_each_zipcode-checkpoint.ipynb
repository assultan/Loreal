### Model to find the products allocation for each Retailer per Zip Code

### Product Allocation for London Drugs

```python
df = df.rename(columns={'Door Zip Code': 'ZipCode', 'Retailer Code': 'RetailerCode', 'POS Units': 'Units'})
```

```python
model_df_ld = df.loc[df['Retailer Name'] == 'LONDON DRUGS'][['RetailerCode', 'ZipCode', 'UPC', 'Units']]
model_df_ld.head()
```

```python
unique_upcs = model_df_ld['UPC'].unique()
unique_upcs.shape
```

```python
model_df_ld_allocated = model_df_ld[model_df_ld['Units']>0]
model_df_ld_returned = model_df_ld[model_df_ld['Units']<0]
```

### Prepare the dataset where find the appropriate number of items against each upc for each retailers each Zip code i.e., UPC = items allocated - items returned for each retailer at each UPC i.e UPC 123 is being allocated 10 times to London Drugs and 13 times to Jean-Coutu and returned 7 times from London drugs and 5 times from Jean-Coutu so the desired data format would be:

London Drugs at L1k 0h5 = 10-7 = 3
London Drugs at M8W 1Y7 = 3-4 = -1

Jean Coutu at L1k 0h5 = 4-7 = -3
Jean Coutu at M8W 1Y7 = 10-2 = 8

```python
### UPCs for each zip code
ld_adjusted_products = {}
zips_ld = []
upcs_ld = []
units_ld = []
the_retailers = []
for zipcode in unique_zipcodes:
    allocated_units = model_df_ld_allocated[model_df_ld_allocated['ZipCode'] == zipcode][['ZipCode','UPC', 'Units']]
    returned_units = model_df_ld_returned[model_df_ld_returned['ZipCode'] == zipcode][['ZipCode','UPC', 'Units']]
        
    if (returned_units.shape[0]>0):
        returned_dict = pd.Series(returned_units.Units.values,index=returned_units.UPC).to_dict()
        for k, v in returned_dict.items():
            if k in list(allocated_units['UPC']):
                the_units = allocated_units.loc[(allocated_units['ZipCode'] == zipcode) & (allocated_units['UPC'] == k)]['Units'] + v
                final_units = the_units.values[0] 
                allocated_units.loc[the_units.index[0], 'Units'] = final_units
            else:
                allocated_units.loc[allocated_units.shape[0]+1] = [zipcode, k, v]
    aggregated_data = allocated_units.groupby(['UPC'])['Units'].sum()
    zips_ld.extend([zipcode for x in range(aggregated_data.shape[0])])
    the_retailers.extend(['LondonDrugs' for x in range(aggregated_data.shape[0])])
    upcs_ld.extend(list(aggregated_data.keys()))
    units_ld.extend(list(aggregated_data.values))


#     
```

```python
df_upcs_for_each_retailer_zipcode_ld = pd.DataFrame({'Retailer':the_retailers, 'ZipCode': zips_ld, 
                                                  'UPC': upcs_ld, 'Units': units_ld})
```

### Product Allocation for JEAN-COUTU

```python
model_df_jc = df.loc[df['Retailer Name'] == 'JEAN-COUTU'][['RetailerCode', 'ZipCode', 'UPC', 'Units']]
model_df_jc.head()
```

```python
unique_upcs = model_df_jc['UPC'].unique()
unique_upcs.shape
```

```python
model_df_jc_allocated = model_df_jc[model_df_jc['Units']>0]
model_df_jc_returned = model_df_jc[model_df_jc['Units']<0]
```

```python
model_df_jc_allocated
```

```python
unique_zipcodes = model_df_jc_allocated['ZipCode'].unique()
```

```python
len(unique_zipcodes)
```

```python
### UPCs for each zip code
jc_adjusted_products = {}
zips_jc = []
upcs_jc = []
units_jc = []
the_retailers = []
for zipcode in unique_zipcodes:
    #get the sum of all the products that were allocated for each upc
    allocated_units = model_df_jc_allocated[model_df_jc_allocated['ZipCode'] == zipcode][['ZipCode','UPC', 'Units']]
    returned_units = model_df_jc_returned[model_df_jc_returned['ZipCode'] == zipcode][['ZipCode','UPC', 'Units']]
        
    if (returned_units.shape[0]>0):
        returned_dict = pd.Series(returned_units.Units.values,index=returned_units.UPC).to_dict()
        for k, v in returned_dict.items():
            if k in list(allocated_units['UPC']):
                the_units = allocated_units.loc[(allocated_units['ZipCode'] == zipcode) & (allocated_units['UPC'] == k)]['Units'] + v
                final_units = the_units.values[0] 
                allocated_units.loc[the_units.index[0], 'Units'] = final_units
            else:
                allocated_units.loc[allocated_units.shape[0]+1] = [zipcode, k, v]
                
    aggregated_data = allocated_units.groupby(['UPC'])['Units'].sum()
    zips_jc.extend([zipcode for x in range(aggregated_data.shape[0])])
    the_retailers.extend(['JeanCoutu' for x in range(aggregated_data.shape[0])])
    upcs_jc.extend(list(aggregated_data.keys()))
    units_jc.extend(list(aggregated_data.values))
```

```python
df_upcs_for_each_retailer_zipcode_jc = pd.DataFrame({'Retailer':the_retailers, 'ZipCode': zips_jc, 
                                                  'UPC': upcs_jc, 'Units': units_jc})
```

```python
df_upcs_for_each_retailer_zipcode_jc
```

### Concatenating the London Drugs Sales and Jean Coutu Sales

```python
final_df = pd.concat([df_upcs_for_each_retailer_zipcode_ld, df_upcs_for_each_retailer_zipcode_jc])
```

```python
final_df
```

```python
final_df
```

```python
reatail_dict = {'LondonDrugs':0, 'JeanCoutu': 1}
zip_codes = final_df['ZipCode'].unique()
the_zips = {}
for i in range(len(zip_codes)):
    the_zips[zip_codes[i]] = i 
```

```python
the_zips
```

```python
final_df['retailer'] = final_df['Retailer'].map(reatail_dict)
final_df['zipcode'] = final_df['ZipCode'].map(the_zips)
```

```python
final_df.drop(['Retailer', 'ZipCode'], axis=1, inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X = final_df.drop(['Units'], axis=1)
y = final_df['Units']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, preds)
```
